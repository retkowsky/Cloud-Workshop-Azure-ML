## Chargement fichier CSV Scala

```
%scala 
val diamonds = sqlContext.read.format("csv")
  .option("header", "true")
  .option("inferSchema", "true")
  .load("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv")

display(diamonds)
```

## Chargement fichier CSV Python

```
%python
diamonds = sqlContext.read.format('csv').options(header='true', inferSchema='true').load('/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv')

display(diamonds)
```

## Chargement fichier CSV R

```
%r

library(SparkR)
diamondsDF <- read.df("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", source = "csv", header="true", inferSchema = "true")

display(diamondsDF)
```

## Chargement fichier CSV SQL

```
%sql
CREATE TEMPORARY TABLE diamonds
  USING csv
  OPTIONS (path "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header "true", mode "FAILFAST")
```

```
%sql
select * from diamonds where cut="Premium"
```

```
%sql
select * from diamonds where price > 500 and cut="Premium" order by carat
```

```
%sql
select * from diamonds where price > 500 and cut="Premium" order by depth
```

```
%sql
Drop table diamonds;
```
